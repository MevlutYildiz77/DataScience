```python
conn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server}; SERVER=DESKTOP-0AKH5FA; database=SampleSales; TRUSTED_CONNECTION=yes')
cursor = conn.cursor()
cursor.execute('SELECT * FROM product.stock')
columns = cursor.description 
result = []
for value in cursor.fetchall():
    tmp = {}
    for (index,column) in enumerate(value):
        tmp[columns[index][0]] = column
    result.append(tmp)
print(result)
```

```python
conn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server}; SERVER=DESKTOP-0AKH5FA; database=SampleSales; TRUSTED_CONNECTION=yes')
cursor = conn.cursor()
cursor.execute("""SELECT * 
               FROM INFORMATION_SCHEMA.COLUMNS 
               WHERE TABLE_NAME = N'product.product'""")
columns = cursor.description 
result = []
for value in cursor.fetchall():
    tmp = {}
    for (index,column) in enumerate(value):
        tmp[columns[index][0]] = column
    result.append(tmp)
print(result)
```
